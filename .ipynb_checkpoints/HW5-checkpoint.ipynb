```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load
city_data_to_load = "city_data.csv"
ride_data_to_load = "ride_data.csv"

# Read the City and Ride Data
city_df = pd.read_csv(city_data_to_load)
ride_df = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset

database = pd.merge(ride_df,city_df,on="city",how="left")

# Display the data table for preview

database.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types

urban_data = pd.DataFrame(database[(database["type"]=="Urban")])
suburban_data = pd.DataFrame(database[(database["type"]=="Suburban")])
rural_data = pd.DataFrame(database[(database["type"]=="Rural")])

#Urban data

avg_fare_urban = urban_data.groupby(["city"])["fare"].mean()
total_rides_urban = urban_data.groupby(["city"])["ride_id"].count()
total_drivers_urban = urban_data.groupby(["city"])["driver_count"].value_counts

#Sub Urban data

avg_fare_suburban = suburban_data.groupby(["city"])["fare"].mean()
total_rides_suburban = suburban_data.groupby(["city"])["ride_id"].count()
total_drivers_suburban = suburban_data.groupby(["city"])["driver_count"].value_counts

#Rural data

avg_fare_rural = rural_data.groupby(["city"])["fare"].mean()
total_rides_rural = rural_data.groupby(["city"])["ride_id"].count()
total_drivers_rural = rural_data.groupby(["city"])["driver_count"].value_counts

# Build the scatter plots for each city types

plt.scatter(total_rides_urban,avg_fare_urban,s=total_rides_urban*10,
           marker = "o",facecolors="coral",edgecolors="black",alpha=0.75,label="Urban")
            
plt.scatter(total_rides_suburban,avg_fare_suburban,s=total_rides_suburban*10,
           marker = "o",facecolors="skyblue",edgecolors="black",alpha=0.75,label="Suburban")

plt.scatter(total_rides_rural,avg_fare_rural,s=total_rides_rural*10,
           marker= "o",facecolors="gold",edgecolors="black",alpha=0.75,label="Rural")

# Incorporate the other graph properties

plt.title("Pyber Ride Sharing Data (2016)", fontsize= "12")
plt.legend(loc='upper right', shadow=True)

# Create a legend

plt.xlabel("Total Numbers of Rides (Per City)")
plt.ylabel("Average Fare($)")

# Incorporate a text label regarding circle size

plt.text(42, 40, "Note:")
plt.text(42,38,"Circle size correlates with driver count per city")

# Save Figure

plt.savefig("Pyber_scatter_data.png")
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents

total_fares = database.groupby(["type"])["fare"].sum()
colors = ["gold","skyblue","coral"]
labels = ["Rural","Suburban","Urban"]
explode = (0,0,0.1)

# Build Pie Chart

plt.pie(total_fares,labels=labels,colors=colors,explode=explode,autopct="%1.2f%%", shadow=True,startangle=150)
plt.title("% of Total Fares by City Type",fontweight="bold",fontsize="10")


# Save Figure

plt.savefig("Total_fares_by_citytype.png")
plt.show()
```

```python
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents

total_rides = database.groupby(["type"])["ride_id"].count()
colors = ["gold","skyblue","coral"]
labels = ["Rural","Suburban","Urban"]
explode = (0,0,0.1)

# Build Pie Chart

plt.pie(total_rides,labels=labels,colors=colors,explode=explode,autopct="%1.2f%%",shadow=True,startangle=150)
plt.title("% of Total Rides by City Type",fontweight="bold",fontsize="10")

# Save Figure

plt.savefig("Total_rides_by_citytype.png")
plt.show()
```

```python
# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents

data_driver = database.drop_duplicates('city')
total_driver = data_driver.groupby(["type"])["driver_count"].sum()
colors = ["gold","skyblue","coral"]
labels = ["Rural","Suburban","Urban"]
explode = (0,0,0.1)

# Build Pie Charts

plt.pie(total_driver,labels=labels,colors=colors,explode=explode,autopct="%1.2f%%",shadow=True,startangle=150)
plt.title("% of Total Drivers by City Type",fontweight="bold",fontsize="10")

# Save Figure

plt.savefig("Total_drivers_by_citytype.png")
plt.show()
```

```python
# Show Figure
plt.show()
```
